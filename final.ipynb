```python
G = nx.Graph()

# 5 node star
G.add_nodes_from(["A", "B", "C", "D", "E"])

G = G.to_directed()

G.add_edge("B", "A")
G.add_edge("C", "A")
G.add_edge("D", "A")
G.add_edge("E", "A")

nx.draw_networkx(G, node_color='black', font_color='white', edge_color='black', font_weight='bold',  node_size=700)
plt.axis('off')
plt.show()
```
