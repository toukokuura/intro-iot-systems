**(1) Analyze an electric heater as a system following Example 1.1.**

As per the system engineering definition, a system is a set of components with attributes, and with relationships to each other to achieve some common purpose. The purpose of the electric heater would be to heat the space that it is in, likely to a specific temperature. 

Components can be divided to structural, operating and flow components. For the heater, examples of structural components are the outer shell of the heater, pipes that hold air/liquid, or mounting brackets. Operating components would be a thermostat, transformers, fans, motors etc. Flow components would be electricity, which is used as the resource to generate the heat.
